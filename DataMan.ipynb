```python
tempdf=pd.DataFrame({
    'city':['noida','mumbai','delhi','lko'],
    'temp':[35,40,45,38],},index=[0,1,2,3])
tempdf
```

```python
winddf=pd.DataFrame({                           #index use to allign column head
    'city':['lko','noida','delhi','mumbai'],
    'winds':[5,4,5,8],},index=[3,0,1,2])
winddf
```

```python
df=pd.concat([tempdf,winddf],axis=1) 
df
```

```python
tempdf=pd.DataFrame({
    'city':['noida','mumbai','delhi','lko','agra'],
    'temp':[35,40,45,38,32],})
tempdf
```

```python
winddf=pd.DataFrame({                           #index use to allign column head
    'city':['lko','noida','delhi','mumbai'],
    'winds':[5,4,5,8],})
winddf
```

```python
df=pd.merge(tempdf,winddf,on='city',how='outer',indicator=True) #if city will be diff than it will show d common city info
df                                              #outer will do union  and left take temp side  and right take wind and common
```

```python
tempdf1=pd.DataFrame({
    'city':['noida','mumbai','delhi','lko','agra'],
    'temp':[35,40,45,38,32],})
tempdf1
```

```python
tempdf2=pd.DataFrame({
    'city':['noida','mumbai','delhi','lko','agra'],
    'temp':[35,40,45,38,32],})
tempdf2
```

```python
df2=pd.merge(tempdf1,tempdf2,on='city')
df2
```

```python
df2=pd.merge(tempdf1,tempdf2,on='city',suffixes=('left'))
df2
```
